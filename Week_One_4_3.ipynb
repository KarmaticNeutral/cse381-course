## Algo 1: Ascending

```python
def algo1(A, B):
    if A < B * (B + 1) / 2:
        return "Expected Error";
    parts = [];
    sum = 0
    for i in range(1, B):
        parts.append(i);
        sum += i;
    parts.append(A - sum)
    return parts
```

## Algo 5: Descending

```python
def algo5(A, B):
    if A < B * (B + 1) / 2:
        return "Expected Error";
    parts = []
    remainder = A
    next = B - 1
    while len(parts) < B - 1:
        parts.append(next)
        remainder = remainder - next
        next = next - 1
    parts.append(remainder)
    return parts
```

## Algo 3: Square Root

```python
from math import floor, sqrt
def algo3(A, B):
    if A < B * (B + 1) / 2:
        return "Expected Error";
    parts = [];
    remainder = A;
    while len(parts) < B - 1:
        if remainder < 0:
            return "Subzero Remainder While Running"
        bit = floor(sqrt(remainder));
        i = 1;
        while bit in parts:
            if bit + i not in parts:
                bit = bit + i;
            if bit - 1 not in parts:
                bit = bit - i;
            i = i + 1;
        if bit < 0:
            print("Subzero Error!", A, B);
        parts.append(bit);
        remainder = remainder - bit;
    parts.append(remainder);
    return parts;
```
